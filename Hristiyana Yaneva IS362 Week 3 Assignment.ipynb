# Bicycle Miles Data
This notebook tracks the total miles ridden each day and calculates the daily miles based on the cumulative miles.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Making the pandas series, can keep adding total mile entries here

cumulative_miles = pd.Series([55, 120, 180, 250, 330], name = 'Total Miles' )
```

## Making the Pandas DataFrame

The index is set to start at 1 instead of 0. 

```python
df = pd.DataFrame(cumulative_miles)
df.index = range(1, len(df) + 1)
df.index.name = "Day"
```

```python
df['Daily Miles'] = df['Total Miles'].diff().fillna(df['Total Miles'].iloc[0])
```

## Displaying the DataFrame


```python
display(df.style.set_caption("Bicycle Trip Summary").format({'Daily Miles': '{:.0f}'}))
```

## Stacked bars chart visualization

Making a simple stacked bar chart to represent the total miles, as well as the daily miles ridden. 
This kind of chart might not be ideal for long-term use because the total miles will keep increasing, making it difficult to see the logs from the earliest days where the number was not as high.

```python
df.plot(kind='bar', y=['Total Miles', 'Daily Miles'], stacked=True, figsize=(11, 7), color=['lightgreen', 'skyblue'])
plt.xlabel("Day")
plt.ylabel("Miles")
plt.title("Total vs. Daily Miles")
plt.legend(["Total Miles", "Daily Miles"])
plt.xticks(rotation=0) #Makes the number of days vertical
plt.show()
```

## Line chart visualization

A line chart, such as the one below, might be more appropriate for tracking for long-term use.

```python
plt.figure(figsize=(11, 7))

plt.plot(df.index, df["Total Miles"], marker='o', linestyle='-', color='red', label="Total Miles")
plt.plot(df.index, df["Daily Miles"], marker='.', linestyle='--', color='blue', label="Daily Miles")

plt.xlabel("Day")
plt.ylabel("Miles")
plt.title("Total and Daily Miles Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
